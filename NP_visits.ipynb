```python
import pandas as pd
import requests

# get data from url
url = 'https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States'
html = requests.get(url).text

# find table and clean and drop columns and American Samoa row
table = pd.read_html(html)
np = table[0]
np.drop(columns=['Image','Description'],inplace=True)
np = np[~np['Name'].str.contains('American Samoa')]

np['Date established as park (revisions)'] = 'revise ' + np['Date established as park[12]'] 
np['Unique Row ID'] = range(1, len(np) + 1)
np.set_index('Unique Row ID',inplace=True)
np.to_csv('np_wiki.csv')
np.head()
```
